```python
from google.colab import files
uploaded = files.upload()
csv_file = '2025제주평균기온.csv'
```

2. 업로드한 파일을 판다스를 사용하여 데이터 프레임(df)에 저장하고 출력한다.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(csv_file, encoding ='cp949')
print(df)
```

```python
df['일자'] = pd.to_datetime(df['일자'])
df['연도'] = df['일자'].dt.year
df['월'] = df['일자'].dt.month
df['일'] = df['일자'].dt.day
```

```python
df
```

3. 날짜별 평균 기온 막대그래프로 표현한다.



```python
#날짜별 평균 기온 막대그래프로 그리기
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 코랩용 한글 폰트 적용
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False

plt.subplot(1,3,1)
df_sorted = df.sort_values(by='평균기온',ascending=True)

plt.barh(df_sorted['일자'], df_sorted['평균기온'], color='skyblue', edgecolor='black')

plt.title('제주 평균기온', fontsize=14)
plt.xlabel('평균기온', fontsize=12)
plt.ylabel('일자', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
