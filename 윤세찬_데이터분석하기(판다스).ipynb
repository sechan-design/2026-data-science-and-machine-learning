## **① Series 객체 생성하기**
Pandas의 Series 클래스에서는 Numpy의 ndarray 1차원 배열을 기본으로 정수와 문자열 인덱스를 사용할 수 있습니다.




````python
#(예제) Series 객체 생성하기
import pandas as pd          # pandas 패키지 가져오기(일반적으로 네임페이스를 pd로 사용)

obj = pd.Series([5,-4,7,0,12])  # pd.Series() 메소드는 1차원 배열의 인덱스와 값으로 생성. Series 객체의 기본 형(Type)은 64비트 정수형
print(obj)
````

## **② Series 객체의 값과 인덱스**
Series 객체에서 값(Value)과 인덱스(index)를 얻기 위해서는 각각 Series 클래스의 value와 index 속성을 이용한다.

````python
#(예제) Series 객체의 값과 인덱스
#출력값과 동일한 값이 출력되도록 코드를 완성해봅시다.
import pandas as pd

obj = pd.Series([8,-20,-3,13,2])    # Series 객체 생성

print(obj.values)                   # values속성을 이용하여 Series 객체의 값을 출력
print(obj.index)                    # index 속성을 이용하여 Series 객체의 인덱스를 출력

print(obj[0:5])                       #Series 객체의 특정 인덱스(3번째 인덱스) 출력
````

## **③ Series 객체에 인덱스 설정하기**
Series 객체에서는 0부터 시작하는 숫자 인덱스 외에 문자열 인덱스를 설정할 수 있다.

````python
#(예제) Series 객체에 인덱스 설정
#출력값과 동일한 값이 출력되도록 코드를 완성해봅시다.
import pandas as pd

obj = pd.Series([10,20,30,40,50], index=['a','b','c','d','e']) #인덱스를 a,b,c,d,e로 설정해봅시다.
print(obj)

print(obj['c'])       # 특정 인덱스 'c'에 해당되는 요소 값 출력
print(obj[['a','d']])     # 특정 인덱스 'd', 'a'에 해당되는 요소 값 출력
print(obj[:])       # 슬라이싱을 이용하여 특정 인덱스 1에서 3까지의 요소 값 출력
````

##④ Series 객체의 연산과 조건식

````python
#(예제) Series 객체의 연산과 조건식
import pandas as pd

obj = pd.Series([10,20,30,40,50])

print(obj*10)         #객체의 모든 요소 값에 10 곱한 값 반환

print(obj[obj>39])    #obj 객체의 요소 중 값이 25보다 큰 요소 값을 반환
````

##⑤ Series 객체와 for문

**💡Tip! pandas.Series.items() 메소드**

Series 클래스의 items() 메소드는 Series 객체에 있는 요소를 튜플 형태, 즉 (index, value)로 반환.

````python
#(예제) Series 객체와 for문
import pandas as pd

pop = pd.Series([9765623,3441453,2461769],index=['서울','부산','대구'])

for i,v in pop.items() :        #pop.items()는 Series 객체 pop의 인덱스와 값을 얻는 데 사용
  print('%s : %d명' % (i,v))    #pandas.Series.items() 메소드는 Series 객체에 있는 요소를 튜플 형태, 즉 (index, value)로 반환. i는 객체의 index가 저장되고, v에는 객체의 요소 값이 저장.
````

##⑥ Series 객체와 딕셔너리
딕셔너리는 Series 객체와 생김새가 유사하기 때문에 딕셔너리를 이용하여 Series 객체를 생성하는 경우가 많습니다.


⚠️ dictionary(딕셔너리)
딕셔너리는 _______(key)와 _______(value)이 한 쌍이 _______의 대응 관계를 가지고 있는 자료형.

딕셔너리는 데이터를 특정 _______와 _______을 한쌍으로 이루어 순서와 상관없이 **키를 이용해서 바로 값에 접근이 가능**하도록 만든 데이터 타입.


|일자|확진자수|
|:------|:---:|
|서울|9765623|
|부산|3441453|
|대구|2461769|
|광주||
|대전||

````python
#6 Series 객체의 딕셔너리 사용
import pandas as pd

#딕셔너리를 이용하여 서울, 부산, 대구, 광주, 대전의 값을 입력하세요.
pop = pd.Series({'서울': 9765623,'부산': 3441453,'대구': 2461769}, index=['서울','부산','대구','광주','대전'])

print(pop)              # 딕셔너리를 이용하여 생성되는 Series 객체의 형(Type)은 64비트 실수형. 데이터  값이 없는 '광주'와 '대전'은 NaN으로 표시.

pop['광주'] = 149336    # 특정 인덱스 요소에 값 저장.

print('광주시 인구 : %.0f명' % pop['광주'])
````

##① DataFrame 생성하기
데이터 프레임으로 만들 data를 인자로 넘겨서 데이터 프레임을 생성합니다.


💡Tip!

데이터프레임명 = pd.DataFrame(딕셔너리명, columns = 리스트, index = 리스트)

데이터프레임명에 대해 열 인덱스를 설정하는 데는 columns를 사용하고, 행 인덱스를 설정하는 데는 index를 사용한다.  

````python
#딕셔너리 형태의 데이터 프레임 생성하기
#DataFrame의 키에 해당하는 값이 데이터프레임의 컬럼명으로 설정됩니다.
#여러 개의 컬럼을 생성할 경우, 길이가 모두 동일한 값이 전달됩니다.

import pandas as pd #판다스 추가 코드 입력하기

data = {'학번':range(2000,2010), '성적' :[85,95,75, 70, 100, 100, 95, 85, 80, 85]}
df = pd.DataFrame(data) #데이터프레임 생성 코드 입력하기
df
````

````python
#columns에 인자를 넘겨서 컬럼명의 순서를 바꾸어 생성할 수도 있습니다.

import pandas as pd

#columns = ['성적','학번'] 속성을 추가해봅시다.

data = {'학번':range(2000,2010), '성적' :[85,95,75, 70, 100, 100, 95, 85, 80, 85]}
df = pd.DataFrame(data, columns = ['성적','학번'])
df
````

````python
#데이터프레임을 ndArray와 다르게 인덱스를 특정값으로 직접 설정할 수 있습니다.
#index 인자에 설정할 인덱스 값을 리스트로 만들어서 데이터 프레임을 생성해봅시다.

index = ['a','b','c','d', 'e','f','g','h','i','j']
df = pd.DataFrame(data, index=index)
df
````

###💠loc[ ]
인덱스로 값을 선택하는 기능

* loc은 인덱스값으로 로우를 선택할 수 있다.

* 데이터프레임명에 대해 행 인덱스의 Name 속성이 행_인덱스명, 행_인덱스명,...이고,
     
    열 인덱스의 Name 속성이 열_인덱스명, 열_인덱스명, ..인 요소들로 구성된 DataFrame 객체를 얻는다.

```
데이터프레임명.loc[[행_인덱스명, 행_인덱스명,...],[열_인덱스명, 열_인덱스명, ...]]
```


````python
#(예제)loc을 이용한 DataFrame 요소 추출방법
#코드를 실행하여 결과를 확인해봅시다.

import pandas as pd

data = {'학교명':['가나고', '다라고', '마바고', '사아고', '자차고'],    # 딕셔너리 data 생성
        '학급수':[25,23,15,19,10],
        '학생수':[620,600,550,580,400],
        '교사수':[80,95,70,90,65]}

frame = pd.DataFrame(data,index=['01','02','03','04','05'])             # 딕셔너리 data를 이용하여 DataFrame 객체 frame 생성
print(frame)


print('\n[2행의 학생수]')
print("학생수 : ", frame.loc['02','학생수'])                                         # 행 인덱스 '02', 열 인덱스 '학생 수' 출력

print('\n[4행의 학교명, 학급수, 교사수]')
print(frame.loc['04',['학교명','학급수','교사수']])                     # 행 인덱스 '04', 열 인덱스 '학교명', '학급수', '교사수' 출력
````

###💠iloc[]

* loc에서 Name 속성의 인덱스명을 사용하는 것과 달리 iloc에서는 0으로 시작하는 정수를 인덱스로 사용한다.
* 행_인덱스번호와 열_인덱스번호는 [0], [1:3], [:3], [5:],[:] 등에서와 같이 문자열과 리스트에서 사용하는 인덱스 방식과 거의 유사하다.

```
데이터프레임명.iloc[행_인덱스번호, 열_인덱스번호]
```

````python
#(예제)iloc을 이용한 DataFrame 요소 추출방법
import pandas as pd

data = {'아이디' : ['kim','song','han','choi'],                     # 딕셔너리 data 생성
        '구매상품' : ['상품A','상품B','상품C','상품D'],
        '가격' : [15000,23000,33000,50000],
        '개수' : [3,5,1,10],
        '구매일' : ['0303','0810','0120','0601']}

frame = pd.DataFrame(data)                                          # 딕셔너리를 활용하여 frame 생성
print(frame)


print('\n[2행 0열 요소 출력]]')
print(frame.iloc[2,0])                                              # frame 객체에서 행 인덱스가 2이고 열 인덱스가 0인 요소 값 출력

print('\n[3행 0과 1열 요소 출력]')
print(frame.iloc[3,:2])                                              # frame 객체에서 행 인덱스가 3이고 열 인덱스는 0과 1인 열에 해당되는 요소를 구성하여 객체 출력

print('\n[행 전체에 대해, 열 인덱스가 0과 4인 요소들로 구성된 객체 출력]')
print(frame.iloc[:,[0,4]])                                       # frame 객체에서 행 인덱스 : 행 전체에 대해, 열 인덱스가 0과 4인 요소들로 구성된 객체 출력
````

##⑥ 결측치 처리

* 수집된 데이터에 누락된 값(비어있는 값)이 존재할 경우 해당 데이터를 삭제하거나 다른 값을 채워주는 작업입니다.
* 데이터 분석에서 비어있는 값이 존재할 경우 에러를 발생하며 코드가 실행되지 않으므로, 데이터 분석 이전에 반드시 처리해주어야 합니다.

* 결측치가 차지하는 비율에 따라 처리 방법을 달리합니다.

|결측치 비율 | 처리방법|
|------------|----------|
|10%미만 | 로우 삭제 또는 치환|
|10~50% | 모델 기반 처리|
|50%이상|컬럼 삭제|

####💠결측치 처리함수

* 결측치를 치환할 때에는 replace()나 함수를 활용하여 값을 처리할 수 있습니다.
* 데이터 종류에 따라 다르지만 결측치가 포함된 컬럼(column)의 최대값, 평균값, 중앙값, 최소값, 최빈값, 이전 행값, 지정 값으로 치환할 수 있습니다.


|결측치 처리함수 | 설명|
|------------|----------|
|isna() | 데이터프레임에 결측치가 포함되어 있을 경우 True, 그렇지 않을 경우 False를 반환|
|fillna(치환할 값, inplace=True)| 결측치를 지정한 값으로 치환|
|dropna()|결측치가 포함된 행을 삭제|
