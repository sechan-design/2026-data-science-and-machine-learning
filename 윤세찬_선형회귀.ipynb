```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = {'x(운동횟수)':[1,2,3,4,5,6,7],'y(행복 호르몬 양)':[1,3,8,14,21,29,41]}
df = pd.DataFrame(data)
df
```

```python
x=np.array(df['x(운동횟수)']).reshape(-1,1)
y=np.array(df['y(행복 호르몬 양)']).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(x,y)
print("기울기:",model.coef_,"절편:",model.intercept_)
```

```python
import koreanize_matplotlib
plt.xlabel("운동횟수")
plt.ylabel("행복 호르몬 양")
plt.scatter(x,y)
b0 = model.intercept_[0]
b1 = model.coef_[0]
plt.plot(b0+b1*x,'r')
```

```python
from scipy import stats

#피어슨 상관계수
print({stats.pearsonr(df['x(운동횟수)'],df['y(행복 호르몬 양)'])})
# 스피어먼 상관계수
print({stats.spearmanr(df['x(운동횟수)'],df['y(행복 호르몬 양)'])})
```

# ***해석***

운동횟수와 행복 호르몬 양 사이의 상관관계는 산점도로 보았을 때 양의 상관관계가 나타난다는 것을 알 수 있고 따라서 기울기도 양의 값인 6.607 이 나온다는 것을 알 수 있다.
피어슨 상관계수를 살펴봤을 때 둘 사이의 상관관계가 0.97로 강한 상관관계가 나타난다는 것을 알 수 있고, p-value의 값이 0.00013 정도로 굉장히 낮기 때문에 유의확률도 굉장히 낮다는 것을 알 수 있다.
