```python
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib

X = np.array( [[1,1], [2,1], [1,2], [2,2], [5,4], [6,4], [5,5], [6,5]] )
K = 2
random_indices=np.random.choice(len(X),K,replace=False)
centroids = X[random_indices].copy()
```

```python
X[:,0]
```

```python
X[:,1]
```

```python
colors=["red",'blue']
plt.scatter(X[:,0],X[:,1],c='gray', label="데이터")
plt.scatter(centroids[0,0],centroids[0,1],s=250,c= 'orange',marker='*', label='중심점1')
plt.scatter(centroids[1,0],centroids[1,1],s=250,c='purple',marker='*', label='중심점2')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
labels = []
for point in X:
  dist_to_c1 =np.sqrt((point[0]-centroids[0,0])**2+(point[1]-centroids[0,1]))
  dist_to_c2 =np.sqrt((point[0]-centroids[1,0])**2+(point[1]-centroids[1,1]))

  if dist_to_c1 < dist_to_c2:
    labels.append(0)
  else:
    labels.append(1)
labels = np.array(labels)

for i in range(len(X)):
  plt.scatter(X[i,0],X[i,1],c= colors[labels[i]])
plt.scatter(centroids[0,0],centroids[0,1],s=250,c='red',marker='*', label='중심점1')
plt.scatter(centroids[1,0],centroids[1,1],s=250,c='blue',marker='*', label='중심점2')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
new_centroids = np.zeros((2,2))

group_0_points = X[labels==0]
new_centroids[0] = np.mean(group_0_points,axis=0)

group_1_points = X[labels==1]
new_centroids[1] = np.mean(group_1_points,axis=0)

for i in range(len(X)):
  plt.scatter(X[i,0],X[i,1],c= colors[labels[i]])
plt.scatter(centroids[0,0],centroids[0,1],s=250,c='red',marker='*', label='중심점1')
plt.scatter(centroids[1,0],centroids[1,1],s=250,c='blue',marker='*', label='중심점2')
plt.scatter(new_centroids[0,0],new_centroids[0,1],s=250,c='orange',marker='*', label='중심1')
plt.scatter(new_centroids[1,0],new_centroids[1,1],s=250,c='purple',marker='*', label='중심2')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
