```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn import datasets

n_samples = 500
seed = 30

noisy_moons = datasets.make_moons(
    n_samples=n_samples,
    random_state=seed,
    noise=0.05
)

arr=np.array(noisy_moons[0])
label = np.array(noisy_moons[1])
arr.shape
label.shape
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=seed, n_init="auto")
kmean_label = kmeans.fit_predict(arr)
```

```python
import matplotlib.pyplot as plt

plt.scatter(arr[:,0],arr[:,1],c=kmean_label)
plt.show()
```

```python
dbscan = DBSCAN(eps=0.09 , min_samples=2)
dbscan_label = dbscan.fit_predict(arr)

dbscan_label
```

```python
plt.scatter(arr[:,0],arr[:,1],c=dbscan_label)
plt.show()
```
