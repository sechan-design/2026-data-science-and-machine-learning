```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv(csv_file, encoding='utf-8')
df['국내총생산'] = df['국내총생산'].str.replace(',', '').astype(float).astype(int)
df['소비자물가 총지수'] = df['소비자물가 총지수'].astype(float).astype(int)
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x=np.array(df['국내총생산']).reshape(-1,1)
y=np.array(df['소비자물가 총지수']).reshape(-1,1)
```

```python
model = LinearRegression().fit(x,y)
print("기울기:",model.coef_,"절편:",model.intercept_)
```

```python
from scipy import stats
```

```python
# 피어슨 상관계수
print({stats.pearsonr(df['국내총생산'],df['소비자물가 총지수'])})
# 스피어먼 상관계수
print({stats.spearmanr(df['국내총생산'],df['소비자물가 총지수'])})
```
