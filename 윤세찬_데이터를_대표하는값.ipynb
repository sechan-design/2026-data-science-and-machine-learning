### [1단계] Python으로 평균 구하기


1. 내장 함수 sum(), len()을 사용하여 평균을 구한다.



```python
import pandas as pd

data=[1, 2, 3, 4, 5]
avg=sum(data)/len(data)
print(f"총합: {sum(data)},길이:{len(data)}")
print(f"평균: {avg}")
```

2. 넘파이(numpy) 라이브러리를 사용하여 평균을 구한다.
* numpy의 배열 생성은 array()함수의 인자로 리스트를 넣어 생성할 수 있다.
* 리스트는 다양한 자료형을 저장할 수 있지만 numpy의 array는 같은 자료형만 포함해야 한다.

```python
import numpy as np

a=np.array([1,2,3,4,5])
avg=np.mean(a)
print(f"평균: {avg}")
```

3. 넘파이(numpy)와 랜덤(random) 라이브러리를 사용하여 평균을 구한다.

```python
import numpy as np
import random as rd

li=[]
for i in range(5):
    # append(): 리스트의 마지막에 원소를 추가하는 함수 # randint(1,10): 1 이상 10 이하의 정수(int)를 리턴
    # 5개의 저장된 숫자 출력
    li.append(rd.randint(1,10))
print(f"{li}")
# arr변수에 array()함수를 사용하여 배열 생성
# 5개의 저장된 숫자 출력
arr = np.array(li)
print(f"{arr}")

# avg변수에 mean()함수를 사용하여 평균 반환
# avg = np.mean(arr) # avg = arr.mean()과 결과는 동일함.
```

### [2단계] 판다스 라이브러리로 평균값, 중앙값, 최빈값 구하기
1. 구글 코랩의 files 라이브러리를 사용하여 2019년 3월 대구광역시 평균 기온의 데이터세트를 업로드한다.

```python
from google.colab import files
uploaded = files.upload() # 파일 업로드: 2025제주평균기온.csv


csv_file = '2025제주평균기온.csv'
```

2. 업로드한 데이터세트를 불러온 다음 판다스 라이브러리를 사용하여 3월 대구광역시 평균 기온 목록을 출력한다.

```python
import pandas as pd

#csv파일을 데이터프레임으로 만들기
df = pd.read_csv(csv_file, encoding ="euc-kr")
print(df)
```

3. 판다스 라이브러리를 사용하여 3월 대구광역시 평균 기온의 평균값을 구한다.

```python
#평균값 출력
print(f"평균기온:{df["평균기온"].mean()}")
```

4. 판다스 라이브러리를 사용하여 3월 대구광역시 평균 기온의 중앙값을 구한다.

```python
# quantile() 함수 크기 순서로 나열한 자료를 4등분하는 관측 값으로 정규분포를 크게 벗어나거나 산포(흩어짐)이 큰상황에서는 사분위수가 대푯값으로 사용
print(f"중앙값:{df["평균기온"].quantile()}")
```

5. 판다스 라이브러리를 사용하여 3월 대구광역시 평균 기온의 최빈값을 구한다.

```python
# mode(): 데이터 프레임의 열 데이터 셋에서 최빈값을 반환
print(f"최빈값:{df["평균기온"].mode()}")
# pandas의 value_counts() 함수는 인자로 전달된 데이터 셋의 도수분포표를 반환
print(f"{pd.value_counts(df['평균기온'])}")
```
