```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 코랩용 한글 폰트 적용
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
from google.colab import files
uploaded = files.upload()
csv_file = '사교육참여율'
```

```python
csv_file = '사교육참여율.csv'
par = pd.read_csv(csv_file, encoding ='utf-8')
par
```

```python
par['고등학교'] = par['고등학교'].astype(int)
```

```python
# 학령인구수 데이터 시각화(그래프 그리기)

plt.figure(figsize=(12, 6))
plt.bar(par['연도'], par['고등학교'], color='skyblue', edgecolor='black')

plt.title('연도별 고등학생 사교육 참여율', fontsize=16)
plt.xlabel('연도', fontsize=14)
plt.ylabel('고등학생 사교육 참여율(%)', fontsize=14)
plt.xticks(par['연도'], rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
csv_file = '월평균 사교육비.csv'
money = pd.read_csv(csv_file, encoding ='cp949')
money
```

```python
#월평균 사교육비 데이터 시각화(그래프 그리기)
plt.figure(figsize=(12, 7))

plt.plot(money['연도'], money['고등학교'], marker='o', label='고등학교')

plt.title('연도별 월평균 고등학생 사교육비', fontsize=16)
plt.xlabel('연도', fontsize=14)
plt.ylabel('월평균 고등학생 사교육비 (만원)', fontsize=14)
plt.xticks(money['연도'], rotation=45, ha='right')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
