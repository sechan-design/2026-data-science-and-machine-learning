### (활동1) 데이터를 산점도로 나타내기


넘파이(Numpy)의 random 서브패키지로 난수를 생성하는 함수에 대해 알아보자


*   rand: 0부터 1사이의 균일 분포
*   randn: 표준 정규 분포
*   randint: 균일 분포의 정수 난수






1. 넘파이(Numpy)의 random 서브 패키지로 난수를 생성하는 함수에 대해 알아본다.

```python
import numpy as np
import matplotlib.pyplot as plt
#시드 설정
np.random.seed(0)
N=5
x=np.random.rand(N)
# 0부터 1사이에서 균일한 확률 분포로 5개의 실수 난수 생성
print(x)
```

2. 균일한 확률 분포로 50개의 실수 난수와 산점도를 표현한다.

```python
import numpy as np
import matplotlib.pyplot as plt
#시드 설정
#np.random.seed(0)
# 0부터 1사이에서 균일한 확률 분포로 50개의 실수 난수 생성
N=50
# x, y 좌표
x=np.random.rand(N)
y=np.random.rand(N)
colors = np.random.rand(N)
 # The marker size in points**2, 0 to 15 point radii
area=(30*np.random.rand(N))**2

plt.scatter(x,y,c=colors,s=area,alpha=0.5)
```

### (활동2) 피어슨 상관관계 구하기


1. scipy 패키지의 stats 모듈 내의 pearsonr 함수와 spearmanr 함수를 이용하여 상
관계수를 구한다.



```python
import numpy as np
from scipy import stats

weight=np.array([1,2,3,4,5,6,7,8]) # 상품의 무게
price=np.array([0,1,2,2,4,4,5,7]) # 상품의 가격

# 피어슨 상관계수
print({stats.pearsonr(weight,price)})
# 스피어먼 상관계수
print({stats.spearmanr(weight,price)})
```
